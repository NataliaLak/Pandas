```python
taxi.pickups.sum()
```

сгруппируйте данные по району (borough) и укажите, из какого пункта было совершено больше всего поездок.

```python
taxi.groupby('borough', as_index=False).agg({'pickups': 'sum'})
```

```python
taxi.groupby('borough') \
    .agg({'pickups': 'sum'}) \
    .idxmax()
```

Сохраните название района с наименьшим числом поездок в переменную min_pickups, применив подходящий метод.

```python
min_pickups = taxi.groupby('borough') \
    .agg({'pickups': 'sum'}) \
    .idxmin()
```

```python
min_pickups
```

Сгруппируйте данные по двум признакам: району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и выберите районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни.

```python
taxi.groupby(['borough', 'hday']) \
    .agg({'pickups': 'mean'})
```

Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и сохраните результирующий датафрейм в pickups_by_mon_bor.

Обратите внимание, что итоговый датасет должен состоять из 3-х колонок: pickup_month, borough, pickups.

```python
pickups_by_mon_bor = taxi.groupby(['borough', 'pickup_month'], as_index=False) \
                          .agg({'pickups': 'sum'}) \
                          .sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor
```

Напишите функцию temp_to_celcius, которая получает на вход колонку с температурой в °F и возвращает значения, переведённые в градусы Цельсия.

Пример работы:

> taxi['temp'][:5]

0    30.0
1    30.0
3    30.0
4    30.0
5    30.0
> taxi['temp_C'] = temp_to_celcius(taxi['temp'])
> taxi['temp_C'][:5]

0   -1.111111
1   -1.111111
3   -1.111111
4   -1.111111
5   -1.111111

```python
def temp_to_celcius(temp_f):
    return (temp_f - 32) * 5.0 / 9.0
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
taxi['temp_C'][:5]
```
